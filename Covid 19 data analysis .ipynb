# Check Correlations

```python
#Drop Columns for Correlations as Country/Region and WHO Region as it's datatype is Object
```

```python
df.drop(['WHO Region'],axis=1,inplace=True)
```

```python
df.drop(['Date'],axis=1,inplace=True)
```

```python
df.drop(['Country/Region'],axis=1,inplace=True)
```

```python
df.corr()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap = 'coolwarm',center=0)
plt.title("CORRELATION HEATMAP")
plt.show()
```

# New Cases

```python
 abs(df.corr())[['New cases']].style.background_gradient(cmap='Reds')
```

# Count Plot for WHO Regions

```python
plt.figure(figsize=(10,6))
sns.countplot(x='WHO Region',data=df)
plt.title("WHO REGIONS")
plt.show()
```

# Total number of Confirmed, Deaths, Recovered and Active cases around the world

```python
plt.figure(figsize=(8,2))
total_count = ['Confirmed', 'Deaths', 'Recovered', 'Active']
column_sums = df[total_count].sum().plot(kind='bar')
plt.xlabel("Status")
plt.ylabel("Count In Millions")
plt.title("Count of Population Status")
```

# New Deaths and New Recovered Cases

```python
sns.boxplot(df['New deaths'])
plt.title('New Deaths')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(df['New recovered'])
plt.title('New Recovered')
```

# Total Recovered and Active cases on the basis of WHO Region

```python
Region = df.groupby(by='WHO Region').sum()
```

```python
Region['Recovered'].plot(kind='pie',  figsize=(8,5), shadow=True, autopct='%0.1f%%') 
plt.title(' Recovered')
```

```python
Region['Active'].plot(kind='pie', figsize=(8,5), shadow=True, autopct='%0.1f%%') 
plt.title(' Active')
```
